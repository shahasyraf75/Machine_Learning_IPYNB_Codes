```python
import pickle
import pandas as pd

# Load The Pickle
mlr = pickle.load(open('multiple_reg.pkl','rb'))  # r --> read
```

```python
# Final Predictions
california = int(input('Is Startup is in California ? Enter 1 if yes else 0 : '))
florida = int(input('Is Startup is in Florida ? Enter 1 if yes else 0 : '))
newYork = int(input('Is Startup is in New York ? Enter 1 if yes else 0 : '))
r_d = int(input('Enter Research and Development Spend : '))
admin = int(input('Enter Administration Spend : '))
market = int(input('Enter Marketing Spend : '))

user_input = {'california':california , 'florida':florida,
              'newYork':newYork , 'r_d':r_d,
              'admin':admin , 'market':market}

user_data = pd.DataFrame(user_input,index=[0])
user_data
```

```python
predict_profit = mlr.predict(user_data.values)
print("Predicted Profit For Startup :",int(predict_profit[0]))
```
