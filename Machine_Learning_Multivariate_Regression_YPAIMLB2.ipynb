```python
# Load the dataset
import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/yash240990/Python/master/Startups_Data.csv')
data.head()
```

```python
data.info()
```

```python
# Data Preparation
# select dependent and independent variable

data.columns
```

```python
x = ['R&D Spend', 'Administration', 'Marketing Spend', 'State']
x = data[x].values
x
```

```python
y = data.Profit.values
y
```

```python
# OneHotEncoding

from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

transformer = ColumnTransformer(
    transformers = [("OneHot",OneHotEncoder(),[3])],
    remainder = 'passthrough'  # please do not remove other columns which is not specified
)

x = transformer.fit_transform(x)
x = x.astype('int')
```

```python
x
```

```python
#  california    florida   newYork    r&d   admin   marketing
#    0             0          1     165349  136897  471784
#    1             0          0     162597  151377  443898
```

```python
# Split the dataset into 2 parts
# a.) Training Set   : apply a ML algo on Training Set
# b.) Testing Set    : we will be using in model evaluation

# 0.2  20% of the data will be used for Testing Purposes
# and remaining 80% of the data will be used for Training

from sklearn.model_selection import train_test_split

x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# Original Shape
print(x.shape)
print(y.shape)
print('\nAfter Train Test Spilt\n')
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Model Development
from sklearn.linear_model import LinearRegression
mlr = LinearRegression()
mlr.fit(x_train,y_train)
```

```python
# save our trained model
import pickle
pickle.dump(mlr,open('multiple_reg.pkl','wb'))   # w --> write means create   b --> byte stream
```

```python
pred_values = mlr.predict(x_test)
pred_values
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,pred_values)
print('Score Of Model :',score)

# r2 tells you how close your predicted numbers are to the real numbers overall
# r2 is a regression performance metrics
# r2 coefficient of determination
# r2 is a number between 0 and 1

# > 0.9        Excellent Fit
# 0.7 - 0.9    Good Fit
# 0.5 - 0.7    Moderate Fit
# <0.5         Weak Fit
```
