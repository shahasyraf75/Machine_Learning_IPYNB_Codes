```python
# Supervised ML
# Classification --> Logistic Regression

# Sigmoid Function
import numpy as np
from sklearn.linear_model import LogisticRegression

# step 2 : Create the dataset
X = np.array([[1],[2],[3],[4],[5],[6]])  # hours_studied
y = np.array([0,0,0,1,1,1])              # passed or not

# step 3 : Create and train the model
model = LogisticRegression()
model.fit(X,y)
```

```python
# predict_proba() stands for predict probabilities
y_prob = model.predict_proba(X)[:,1]     # [:,1]  all rows but from column 1
y_prob
```

```python
# Get the slope and intercept
a = model.coef_[0][0]    # slope       coef_ is a 2d array in scikit-learn
b = model.intercept_[0]  # intercept

print(f'a (slope) : {a}')
print(f'b (intercept) : {b}')


# a = 1.12 means the model thinks "hours_studied" strongly affects the chance of passing
# if we had a = 0.2  the effect would be much weaker

# b = -3.9 means a student who studies very few hours has a very low chance of passing.
```

```python
# Compute the score --> if a student studying for 3.5 hours
score = (1.1207)*(3.5)+(-3.9223)
print(score)
```

```python
1 / (1 + 1.00015)

# 0.49  --> 49% probability
# there is only 49% chance the student will pass
# 0.49 < 0.5  --> predict "Fail"
```

```python
# Import all necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
# Load The Dataset
data = pd.read_csv('diabetes.csv')
data.head()
```

```python
data.info()
```

```python
# Missing value calculation
data.isnull().sum()
```

```python
# check for duplicates
data.duplicated().sum()
```

```python
# select independent and dependent variable
x = data.drop('Outcome',axis=1)  # axis = 1 means column
y = data['Outcome']
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(x_train,y_train)
```

```python
# predict_proba() stands for predict probabilities
# predict_proba() returns 2 columns for the negative and positive class
# get the predicted probabilities of the positive label only
y_prob = model.predict_proba(x_test)[:,1]
y_prob
```

```python
# get the predicted values
y_pred = model.predict(x_test)
y_pred
```

```python
evl1 = pd.DataFrame({'Probabilities':y_prob,'Predicted_Values':y_pred})
evl1.head(50)
```

```python
from sklearn.metrics import confusion_matrix , classification_report
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
sns.heatmap(cm,annot=True,fmt='d',cmap='Blues')    # fmt --> d means display decimal / int value
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
print('Classification Report\n')
print(classification_report(y_test,y_pred))
```

```python
# Precision
# When my model says someone has diabetes, can I trust it?
# Imagine your model said 10 people have diabetes,
# but in reality, only 8 of them truly do.
# Precision=TP + FP / TP
# â€‹10/8  â€‹= 0.8
# Meaning: When my model says â€œYesâ€, itâ€™s correct 80% of the time.



# Recall
# Did my model miss any diabetic people?
# Recall= TP+FN / TP
# 8 / 12 ==> 0.67
# Meaning: My model found 67% of all actual diabetic people.
# Ability to find all positives.

# Precision = How correct your positive predictions are
# Recall = How complete your positive predictions are

# The F1-score is a single number that balances both precision and recall
```
